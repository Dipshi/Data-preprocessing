```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("processed_data.csv")
```

```python
df.membership.isnull().sum()
```

```python
#converting nominal data to numeric data 
df['membership'] = df['membership'].map({'Monthly':0, 'Biannual':1, 'Annual':2})
df['MonthlyFee'] = df['MonthlyFee'].map({'1000-2000':0, '2000-3000':1, '3000-5000':2})
df['AnnualFee'] = df['AnnualFee'].map({'3000-5000':0, '5000-8000':1, '8000-12000':2,'12000-15000':3,'>15000':4})
df['BiannualFee'] = df['BiannualFee'].map({'2000-3000':0, '3000-5000':1, '5000-8000':2,'8000-12000':3})

```

```python
# df["membership"].fillna(df.membership.median(), inplace=True)#replace null values with median
df.dropna(subset=['membership'],inplace=True)#drop the records which have nan in membership feild 
```

```python
#for repalcing missing values with mode
# df.membership
df["BiannualFee"].fillna(df.BiannualFee.median(), inplace=True)
# 
df["AnnualFee"].fillna(df.AnnualFee.median(), inplace=True)

df["MonthlyFee"].fillna(df.MonthlyFee.median(), inplace=True)
```

```python
# df.BiannualFee.isnull().sum()
df.membership.isnull().sum()
```

```python
df['MonthlyFee'] = df['MonthlyFee'].map({0:'1000-2000', 1:'2000-3000', 2:'3000-5000'})
df['AnnualFee'] = df['AnnualFee'].map({0:'3000-5000', 1:'5000-8000',2:'8000-12000',3:'12000-15000',4:'>15000'})
df['BiannualFee'] = df['BiannualFee'].map({0:'2000-3000', 1:'3000-5000',2:'5000-8000',3:'8000-12000'})
```
